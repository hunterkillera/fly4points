```python
import re
def replace_continent(cell_value):
    if re.search(r'\bcontinental\b', cell_value, flags=re.IGNORECASE):
        return 1
    else:
        return cell_value
```

```python
#lowercase all country names
countries['departure'] = countries['departure'].str.lower()
countries['arrival'] = countries['arrival'].str.lower()
#replace departure country name and arrival country name with ISO 3166-1 alpha-2 code
#countries['departure'] = countries['departure'].replace(country_name_to_iso_alpha2_code)
countries['arrival'] = countries['arrival'].replace(country_name_to_iso_alpha2_code)
countries['departure'] = countries['departure'].replace(country_name_to_iso_alpha2_code)
#if any rows from continental column contain continental, replace the whole cell to 1
countries['continental'] = countries['continental'].apply(replace_continent)
#keep rows with continental = 1
countries = countries[countries['continental'] == 1]
countries
```

```python
#save countries to csv
countries.to_csv('niger_result.csv', sep=',', header=True, index=True)
countries
```
